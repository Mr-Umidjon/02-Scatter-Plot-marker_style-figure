# Scatter text, legend, grid

```python
import matplotlib.pyplot as plt
```

## Text in Matplotlib in plots


### Title of plot
10. Insert Title of your plot as "Point coordinates"(Chizmaning sarlavhasini "Point coordinates" deb nomlang)

```python
def plot_title(x, y):
    '''
    Parameters: x and y axis points;
    '''
    plt.scatter(x, y)
    plt.title(label='Point coordinates', pad=25, color='r', size=25)
    plt.show()


x = 3
y = 4
plot_title(x, y)
```

### Text in the plot
11. Print text(Full Name) in the plot (chizmada matnni(to'liq ismingizni) chop eting)

```python
def text_in_plot(x, y):
    '''
    Parameters: x and y axis points;
    '''
    plt.scatter(x, y, c='w')
    plt.text(x, y, s='TEXT', color='b', size=25)
    plt.show()


x = 3
y = 4
text_in_plot(x, y)
```

*12*. Inserting coordinates for each points (har bir nuqta uchun kordinatalarni kiriting)

```python
def each_point_style_size_color_cor(x, y, color, style, size):
    '''
    Parameters: x and y axis points list;
    '''
    # Write the code here
    for i in range(len(x)):
        plt.scatter(x[i], y[i], c=color[i], marker=style[i], s=size)
        plt.text(x[i] + 0.2, y[i], s=f"({x[i]}, {y[i]})")
    plt.show()


x = [1, 2, 3, 4, 10]
y = [2, 4, 6, 3, 7]
color = ['#418bbf', "#ff7f0e", '#2ca02c', '#d62728', '#9467bd']
style = ['+', "p", "s", 'v', '*']
size = 100

each_point_style_size_color_cor(x, y, color, style, size)
```

### Name of Axis
13. Insert name of Axises (O'qlarning nomini kiriting)

```python
def x_y_label(x, y):
    '''
    Parameters: x and y axis points;
    '''
    # Write the code here
    plt.scatter(x, y, c='red')
    plt.xlabel(xlabel='X - axis', color='m', size=25, labelpad=25)
    plt.ylabel(ylabel='Y - axis', color='m', size=25, labelpad=25)
    plt.show()


x = 3
y = 4

x_y_label(x, y)
```

### Legend
14. Adding legend(description of data) for your plot (Chizmangizga legend (ma'lumotlar tavsifi) qo'shing)

```python
def legend_of_data(x, y):
    '''
    Parameters: x and y axis points list;
    '''
    # Write the code here
    color = ['c', 'm', 'r']
    labels = ['cyan', 'magenta', 'red']

    for i in range(len(x)):
        plt.scatter(x[i], y[i], c=color[i], label=labels[i])

    plt.legend()
    plt.show()


x = [1, 4, 2]
y = [2, 4, 3]

legend_of_data(x, y)
```

## Grid
15. Adding grid for your plot (Chizmangiz uchun grid(katak) qo'shing)

```python

def plot_grid(x, y, color, style, size):
    '''
    Parameters: x and y axis points list;
    '''
    # Write the code here
    plt.grid()
    for i in range(len(x)):
        plt.scatter(x[i], y[i], c=color[i], s=size, marker=style[i])
        plt.text(x[i] + 0.2, y[i], s=f'({x[i]}, {y[i]})')
    plt.show()


x = [1, 2, 3, 4, 10]
y = [2, 4, 6, 3, 7]
color = ['#418bbf', "#ff7f0e", '#2ca02c', '#d62728', '#9467bd']
style = ['+', "p", "s", 'v', '*']
size = 100

plot_grid(x, y, color, style, size)
```
