```python
x = [1, 3, 7]
y = [8, 3, 2]
size = [200, 400, 600]
color = ['red', 'g', 'b']
```

```python
plt.scatter(x, y, marker='v', s=size, c=color)
plt.show()
```

```python
plt.scatter(x, y, marker='v', s=size, c=color)

plt.scatter(4, 5, marker='x', s=1000, c='c')

plt.show()
```

```python
x = [5, 2, 7, 2]
y = [2, 16, 4, 16]

plt.scatter(x, y)
plt.title(label='Title', loc='center', pad=25, size="xx-large")
plt.show()
```

```python
x = [5, 2, 7, 2]
y = [2, 16, 4, 16]

plt.scatter(x, y)
plt.title(label='Title', loc='center', pad=25, size=25, color='red', style='italic', weight='bold')
plt.show()
```

```python
x = [5, 2, 7, 2]
y = [2, 16, 4, 16]

plt.scatter(x, y)
plt.title(label='Title', pad=25, size=25, color='red')
plt.text(6, 8, s='data point')
plt.show()
```

```python
x = [5, 2, 7, 2]
y = [2, 16, 4, 16]

plt.scatter(x, y)
plt.title(label='Title', pad=25, size=25, color='red')
plt.text(6, 8, s='data point')
plt.show()
```

```python
x = [1, 2, 5, 6, 8]
y = [1, 2, 3, 7, 5]

plt.scatter(x, y, c='red', s=80)
plt.grid()
for i in range(len(x)):
    plt.text(x[i] + 0.1, y[i] - 0.1, s=f'({x[i]}, {y[i]})')
plt.show()
```

```python
x = [1, 2, 5, 6, 8]
y = [1, 2, 3, 7, 5]

plt.scatter(x, y, c='red', s=80)
plt.xlabel(xlabel='X label', size=15, labelpad=25)
plt.ylabel(ylabel='Y label', size=15, labelpad=25)
plt.grid()
for i in range(len(x)):
    plt.text(x[i] + 0.1, y[i] - 0.1, s=f'({x[i]}, {y[i]})')
plt.show()
```

```python
x1 = [1, 2, 5]
y1 = [1, 2, 5]

x2 = [6, 8]
y2 = [7, 5]
```

```python
plt.scatter(x1, y1, marker='s', color='red', s=80, label='dog')

plt.scatter(x2, y2, marker='o', color='blue', s=80, label='cat')

plt.legend()

plt.show()
```

```python
plt.scatter(x1, y1, marker='s', color='red', s=80, label='dog')

plt.scatter(x2, y2, marker='o', color='blue', s=80, label='cat')

plt.legend(ncol=2, loc='upper center')

plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(x1, y1, marker='s', color='red', s=80, label='dog')

plt.scatter(x2, y2, marker='o', color='blue', s=80, label='cat')

plt.legend(ncol=2, loc='upper center')

plt.show()
```

```python
plt.figure(dpi=110)

plt.scatter(x1, y1, marker='s', color='red', s=80, label='dog')

plt.scatter(x2, y2, marker='o', color='blue', s=80, label='cat')

plt.legend(ncol=2, loc='upper center')

plt.show()
```

```python
plt.figure(dpi=110, facecolor='red', edgecolor='cyan')

plt.scatter(x1, y1, marker='s', color='red', s=80, label='dog')

plt.scatter(x2, y2, marker='o', color='blue', s=80, label='cat')

plt.legend(ncol=2, loc='upper center')

plt.show()
```
